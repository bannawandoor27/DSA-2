## Paliandrome & symmetrical

```python
def palian_sym(string):
    if string[::-1] == string:
        print('String is a palindrome')
        
        if len(string) % 2 == 1:
            print('String is not symmetrical')
        else:
            r_half = string[:(len(string) // 2)]
            l_half = string[(len(string) // 2):]
            if r_half == l_half[::-1]:
                print('String is symmetrical')
            else:
                print('String is not symmetrical')
    else:
        print('String is not a palindrome')
```

```python
palian_sym('mllm')
```

## Removing ith character of string

```python
def rmv_ith(string:str,i:int)->str:
    new_str = ''
    for j in range(len(string)):
       if j!=i: 
        new_str+=string[j]

    return new_str    
```

```python
print(rmv_ith('banna',1))
```
